```python
#importing libraries
import datetime
import math
import pandas as pd
import random 
import radar 
import matplotlib.pyplot as plt
from faker import Faker
fake = Faker()
```

```python
def generateData(n):
  listdata = []
  start = datetime.datetime(2019, 8, 1)
  end = datetime.datetime(2019, 8, 30)
  delta = end - start
  for _ in range(n):
    date = radar.random_datetime(start='2019-08-1', stop='2019-08-30').strftime("%Y-%m-%d")
    price = round(random.uniform(900, 1000), 4)
    listdata.append([date, price])
  df = pd.DataFrame(listdata, columns = ['Date', 'Price'])
  df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
  df = df.groupby(by='Date').mean()

  return df
```

```python
df = generateData(20)
```

**Bar chart**

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import calendar

#generating data
months = list(range(1, 13))
sales_quantity = [random.randint(100, 1000) for _ in months]

plt.rcParams['figure.figsize'] = (12, 5)
fig, axis = plt.subplots()
plt.xticks(months, calendar.month_name[1:13], rotation=20)
axis.bar(months, sales_quantity, width = 0.5)
plt.show()
```

```python
#Plotting the horizontal bar chart

plt.rcParams['figure.figsize'] = (12, 6)
fig, ax = plt.subplots()
plt.yticks(months, calendar.month_name[1:13])
ax.barh(months, sales_quantity, height = 0.5)
plt.show
```

**Scatter plot**

```python
#importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
headers_cols = ['age','min_recommended', 'max_recommended',
'may_be_appropriate_min', 'may_be_appropriate_max', 'min_not_recommended',
'max_not_recommended']
sleepDf = pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/hands-on-exploratory-data-analysis-with-python/master/Chapter%202/sleep_vs_age.csv', header=0, names=headers_cols)

sleepDf.head(10)
```

```python
#plotting the scatter plot
sns.set()
plt.rcParams['figure.figsize'] = (8, 6)
plt.scatter(x=sleepDf["age"], y=sleepDf["min_recommended"])
plt.scatter(x=sleepDf["age"], y=sleepDf['max_recommended'])
plt.xlabel('Age of person in Years')
plt.ylabel('Total hours of sleep required')
plt.show()
```

```python
#plotting the scatter plot with markers as dotted lines
plt.rcParams['figure.figsize'] = (8, 6)
plt.scatter(x=sleepDf["age"]/12, y=sleepDf["min_recommended"])
plt.scatter(x=sleepDf["age"]/12, y=sleepDf["max_recommended"])
plt.show()
```

```python
#Plot with dotted lines 
plt.rcParams['figure.figsize'] = (8, 6)
plt.plot(sleepDf['age']/12, sleepDf['min_recommended'], 'r--', label='min_recommended')
plt.plot(sleepDf['age']/12, sleepDf['max_recommended'], 'g--', label='max_recommended')
plt.xlabel('Age(years)')
plt.ylabel('No of hours needed for sleep')
plt.legend()
plt.show()
```

From the above scatter plot, we can see that the age increases, the hours needed to sleep decreases and becomes constant after 18 years.

```python
#loading iris dataset from sea born


df = sns.load_dataset('iris')
df['species'] = df['species'].map({'setosa': 0, "versicolor": 1,
"virginica": 2})

plt.rcParams['figure.figsize'] = (10, 6)

plt.scatter(df['sepal_length'], df['sepal_width'], c=df['species'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal width')
plt.show()
```

**Bubble chart**

```python
#Bubble chart
plt.rcParams['figure.figsize'] = (8, 6)
plt.scatter(df.petal_length, df.petal_width, s = 50 * df.petal_length * df.petal_width, c=df['species'], alpha = 0.3)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'petal_length': [1.4, 1.5, 1.3, 1.5, 4.7, 4.5, 4.9, 5.0, 5.0, 5.0, 5.0, 5.0],
    'petal_width': [0.2, 0.2, 0.2, 0.2, 1.4, 1.5, 1.5, 1.6, 1.6, 1.6, 1.6, 1.6],
    'species': ['setosa', 'setosa', 'setosa', 'setosa', 
                'versicolor', 'versicolor', 'versicolor', 'virginica', 'versicolor', 'versicolor', 'versicolor', 'versicolor']
}

df = pd.DataFrame(data)

# Scatter plot with alpha
plt.figure(figsize=(8, 6))
plt.scatter(df.petal_length, df.petal_width,
            s=50,  # Size of the markers
            c=df.species.astype('category').cat.codes,  # Convert species to numeric for coloring
            alpha=0.1)  # Set transparency
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Scatter Plot of Petal Length vs Petal Width')
plt.colorbar(label='Species (Encoded)')
plt.grid(True)
plt.show()
```
