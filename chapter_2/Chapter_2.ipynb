```python
#importing libraries
import datetime
import math
import pandas as pd
import random 
import radar 
import matplotlib.pyplot as plt
from faker import Faker
fake = Faker()
```

```python
def generateData(n):
  listdata = []
  start = datetime.datetime(2019, 8, 1)
  end = datetime.datetime(2019, 8, 30)
  delta = end - start
  for _ in range(n):
    date = radar.random_datetime(start='2019-08-1', stop='2019-08-30').strftime("%Y-%m-%d")
    price = round(random.uniform(900, 1000), 4)
    listdata.append([date, price])
  df = pd.DataFrame(listdata, columns = ['Date', 'Price'])
  df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
  df = df.groupby(by='Date').mean()

  return df
```

```python
df = generateData(20)
```

```python
#plotting the line chart for the data

plt.rcParams['figure.figsize'] = (20, 5)
plt.plot(df)
```

**Bar chart**

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import calendar

#generating data
months = list(range(1, 13))
sales_quantity = [random.randint(100, 1000) for _ in months]

plt.rcParams['figure.figsize'] = (12, 5)
fig, axis = plt.subplots()
plt.xticks(months, calendar.month_name[1:13], rotation=20)
axis.bar(months, sales_quantity, width = 0.5)
plt.show()
```

```python
#Plotting the horizontal bar chart

plt.rcParams['figure.figsize'] = (12, 6)
fig, ax = plt.subplots()
plt.yticks(months, calendar.month_name[1:13])
ax.barh(months, sales_quantity, height = 0.5)
plt.show
```
